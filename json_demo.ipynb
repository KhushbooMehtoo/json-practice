```python
# Sort_keys :- use to JSON string should be sorted in alphabetical order

import json

sort_alpha={
    "Name":"Khushboo",
    "Age":20,
    "Branch":"Development",
    "City":"Chandigarh"
}

json_sort= json.dumps(sort_alpha, sort_keys=True)

print(json_sort)
```

```python
# separators =use to remove extra spaces
import json

sort_alpha={
    "Name":"Khushboo",
    "Age":20,
    "Branch":"Development",
    "City":"Chandigarh"
}

json_sort= json.dumps(sort_alpha, separators=(",",":"))   #separators = (item_separator, key_value_separator)

print(json_sort)
```
