```python
# write and store in a file

import json

emp={
    "EMP_ID":"001",
    "EMP_NAME":"Khushboo",
    "SALARY":"45000",
    "DEPATMENT":"Back-End",
    "CITY":"Pune"
}

with open('data.json','w') as file:

    json.dump(emp,file,indent=4)
print("JSON data has been stored in 'emp.json'")
```

```python
import json

with open('data.json','r') as atul:

    x=json.load(atul)
    print(x)
```
